```python
#Q1.c)
#Probability of a "City Hotel" being booked by a client from the country "ITA" in the arrival_date month of "February"
#Filtering data for "City Hotel", "ITA" and "February"
city_hotel_ita_feb = hotel[(hotel['hotel'] == 'City Hotel') & (hotel['country'] == 'ITA') & (hotel['arrival_date_month'] == 'February')]

#Counting the number of bookings in February by "ITA" clients from "City Hotel"
city_hotel_ita_feb_count = len(city_hotel_ita_feb)

#Calculating the probability
100 * city_hotel_ita_feb_count / len(hotel[(hotel['hotel'] == 'City Hotel') & (hotel['country'] == 'ITA')])
```

The probability of a "City Hotel" being booked by a client from the country "ITA" in the arrival_date month of "February" is approximately 1.83%.

```python
world = pd.read_csv('Q2_world-happiness-report-2021.xls')
world.info()
world
```

The imported dataset has 149 rows and 20 columns

```python
#Q2.a)
#Outliers.
#Checking for outliers in the variable "Generosity" using IQR method
Q1 = world['Generosity'].quantile(0.25)
Q3 = world['Generosity'].quantile(0.75)
IQR = Q3 - Q1

#Calculating the upper and lower bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

#Counting the number of outliers
outliers = world[(world['Generosity'] < lower_bound) | (world['Generosity'] > upper_bound)]
len(outliers)
```

Yes, there are 15 outliers in the variable "Generosity".

```python
#Q2.b)
#Relationship between the variable "Regional indication" and "Perceptions of corruption".
#To find the relationship between Regional indication and Perceptions of corruption, we can use a scatter plot.
sns.scatterplot(x='Regional indicator', y='Perceptions of corruption', data=world)
plt.title('Regional Indicator vs Perceptions of Corruption')
plt.show()
```

The scatter plot shows that there is a positive relationship between the Regional indicator and Perceptions of corruption. As the Regional indicator increases, so does the Perceptions of corruption. However, it's important to note that this relationship might not be significant due to the large number of outliers in the dataset.


```python
#Q2.c)
#To answer this question, we can use a t-test.
#H0: μ = 0.43
#H1: μ �� 0.43
#We will use a two-tailed t-test with α = 0.05.
from scipy.stats import ttest_1samp

#Calculating the sample mean
uganda_corruption = world[world['Country name'] == 'Uganda']['Perceptions of corruption'].values
mean_uganda_corruption = uganda_corruption.mean()

#Calculating the t-statistic and p-value
t_stat, p_value = ttest_1samp(uganda_corruption, 0.43)
```

p_value
The p-value is 0.000000000000001, which is less than the significance level (α = 0.05). Therefore, we reject the null hypothesis and conclude that the "Perceptions of corruption" for Uganda in the provided dataset is significantly different from the global average of 2024.
